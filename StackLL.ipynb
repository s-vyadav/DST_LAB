```python
class Node:
  def __init__(self,x):
    self.data=x
    self.next=None

class Stack:
  def __init__(self):
    self.top=None

  def push(self):
      x = int(input("Enter element to be insterted: "))
      new = Node(x)
      if self.top is None:
        self.top=new
        self.top.next=None
      else:
        new.next=self.top
        self.top=new

  def pop(self):
    if self.top.next is None:
      print("Popped element=",self.top.data)
      self.top=None
    else:
      temp=self.top
      print("Popped element=",self.top.data)
      self.top=temp.next
      temp=None

  def display(self):
    if self.top is None:
      print("STACK IS EMPTY")
    else:
      temp=self.top
      while temp:
        print(temp.data)
        temp=temp.next
      print("Element at TOP=",self.top.data) #hence temporaray variable

s = Stack()
size = int(input("Enter limit of stack size: "))
ctr = 0
while True:
  print("Choose any one option\n1.Push\n2.Pop\n3.Display\n4.Exit")
  ch = int(input("Enter your choice: "))
  if ch == 1:
    print("Push Operation")
    if ctr<=size-1:
      s.push()
      ctr+=1
    else:
      print("STACK OVERFLOW")
  elif ch == 2:
    print("Pop Operation")
    if s.top is None:
      print("STACK UNDERFLOW")
    else:
      s.pop()
      ctr-=1
  elif ch == 3:
    print("Elements of Stack: ")
    s.display()
  elif ch == 4:
    print('EXIT!')
    break
  else:
    print("Enter valid innput")
```
